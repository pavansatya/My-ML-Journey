```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression
```

These are the regression coefficient A.K.A slope and regression intercepts respectively

```python
reg = LinearRegression()
reg.fit(X,y)

reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
y_pred = ((78.35 * X) + 100).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b1 = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 100

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
# Lets take learning rate = 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python

# Calculating the new intercept
b2 = b - step_size
b2
```

```python
y_pred1 = ((78.35 * X) + b2).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b2 = {}'.format(b2))
plt.plot(X,y_pred,color='#A3E4D7',label='b1 = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b2)
loss_slope
     
```

```python
step_size = loss_slope*lr
step_size
```

```python
b3 = b2 - step_size
b3
   
```

```python
y_pred2 = ((78.35 * X) + b3).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b3 = {}'.format(b3))
plt.plot(X,y_pred1,color='#A3E4D7',label='b2 = {}'.format(b2))
plt.plot(X,y_pred,color='#A3E4D7',label='b1 = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b3)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b4 = b3 - step_size
b4
```

```python
y_pred3 = ((78.35 * X) + b3).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b4 = {}'.format(b4))
plt.plot(X,y_pred2,color='#A3E4D7',label='b3 = {}'.format(b3))
plt.plot(X,y_pred1,color='#A3E4D7',label='b2 = {}'.format(b2))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1 # Increase the learning rate and it will take a lot of steps to converge

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
